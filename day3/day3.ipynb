# DAY 3 : Video Game Sales

```python
import pandas as pd
import numpy as np
df = pd.read_csv("vgsales.csv")
```

## Question 1

```python
print(f"Total Rows = {df.shape[0]}")
print(f"Total Columns = {df.shape[1]}\n")

summary_table = pd.DataFrame({
    "Data Type" : df.dtypes,
    "Missing Values Count" : df.isna().sum()
})

print(summary_table)
```

## Question 2

```python
sorted_df = df.sort_values(by="Global_Sales", ascending=False)
cols = ['Rank', 'Name', 'Platform', 'Global_Sales']
top_10 = sorted_df[cols].iloc[:10]
top_10
```

## Question 3

```python
publisher_analysis = df.groupby("Publisher").agg(Total_Global_Sales = ("Global_Sales","sum"))
top_10_publishers = publisher_analysis.sort_values(by="Total_Global_Sales", ascending=False).iloc[:10]
print(top_10_publishers)
```

## Question 4

```python
platform_analysis = df.groupby("Platform").agg(
    Number_of_Games = ("Name",'count'),
    Total_Global_Sales = ("Global_Sales",'sum')
).reset_index()
platform_analysis = platform_analysis.sort_values(by='Total_Global_Sales', ascending=False)
print(platform_analysis)
```

## Question 5

```python
regional_sales = pd.DataFrame({
    "Region" : ['NA','EU','JP','Other'],
    "Total_Sales" : [df['NA_Sales'].sum(), df["EU_Sales"].sum(), df["JP_Sales"].sum(), df["Other_Sales"].sum()]
})

print(regional_sales)
print("\nNA generates most revenue")
```

## Question 6

```python
genre_performance = df.groupby("Genre").agg(
    Avg_Global_Sales = ("Global_Sales","mean"),
    Total_Global_Sales = ("Global_Sales",'sum')
)
genre_performance.sort_values(by='Total_Global_Sales', ascending=False, inplace=True)

print(genre_performance)
```

## Question 7

```python
platform_based_genre = df.groupby(["Platform","Genre"]).agg(Total_Global_Sales = ("Global_Sales",'sum')).reset_index()
top_15 = platform_based_genre.sort_values(by="Total_Global_Sales", ascending=False)

print(top_15.iloc[:15])
```
