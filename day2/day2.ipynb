# Day 2 : Titanic Dataset

```python
import pandas as pd
import numpy as np

data = pd.read_csv("train.csv")
```

## Question 1

```python
total_females = data.groupby("Sex")['PassengerId'].count().iloc[0]
total_males = data.groupby("Sex")['PassengerId'].count().iloc[1]

survived_females = data.groupby("Sex")['Survived'].sum().iloc[0]
survived_males = data.groupby("Sex")['Survived'].sum().iloc[1]

print(f"1. Total Females = {total_females}")
print(f"2. Total Males = {total_males}")
print(f"3. Survived Females = {survived_females}")
print(f"4. Survived Males = {survived_males}")
```

## Question 2

```python
multi_cond = data[data["Pclass"]==1].groupby("Sex")["Survived"].sum().iloc[0]
print(f"Number of female passengers in Pclass=1 who survived = {multi_cond}")
```

## Question 7

```python
data['Age_Filled'] = data['Age'].fillna(data['Age_Median_By_Class'])
print(f"Missing Before = {data['Age'].isna().sum()}")
print(f"Missing After = {data['Age_Filled'].isna().sum()}")
```

## Question 8

```python
def age_group(x):
    if x<18:
        return "Child"
    elif 18<=x<=59:
        return "Adult"
    elif x>=60:
        return "Senior"

data['Age_Group'] = data['Age_Filled'].apply(age_group)

data.head()
```

## Question 9

```python
data.groupby('Age_Group').agg(
    Passenger_Count = ("PassengerId",'count'),
    Survival_Rate = ("Survived", lambda x: round(x.mean()*100,2))
)
```
