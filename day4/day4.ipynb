# Day 4 : OList Customers Dataset

```python
import pandas as pd
import numpy as np

customers = pd.read_csv("olist_customers_dataset.csv")
orders = pd.read_csv("olist_orders_dataset.csv")
order_items = pd.read_csv("olist_order_items_dataset.csv")
products = pd.read_csv("olist_products_dataset.csv")
```

## Question 1

```python
datasets = [orders,order_items,customers,products]
names = ['orders','order_items','customers','products']
count = 0
summary = pd.DataFrame()
for dataset in datasets:
    summary_dict = {
        "Dataset": names[count],
        "Rows": dataset.shape[0],
        'Columns': dataset.shape[1],
        'Missing values': dataset.isna().sum().sum() ## len(dataset.dropna()) ## Correction
    }
    count+=1
    temp = pd.DataFrame([summary_dict])
    summary = pd.concat([summary,temp], ignore_index=True)


summary

# I took some help for syntax
```

## Question 2

```python
first_merge = pd.merge(orders,customers, on='customer_id', how='left')
first_merge.head()
print(f"1. Rows = {first_merge.shape[0]}")
print(f"2. Row count did not change")
print(f"3. Because the rows in both dfs were the same before, on joining, the it retains all the rows as it is mapped to customer_id")
```

## Question 3

```python
top_10_by_state = first_merge.groupby('customer_state').agg(number_of_customers = ('customer_id','count')).reset_index().sort_values('number_of_customers', ascending=False).iloc[:10]
top_10_by_state
```

## Question 4

```python
revenue = round(order_items['price'].sum(),2)
avg_price = round(order_items['price'].mean(),2)
highest_price = order_items['price'].max()

print(f"1. Total Revenue = {revenue}")
print(f"2. Avg Price = {avg_price}")
print(f'3. Highest Price = {highest_price}')
```

## Question 5

```python
status = orders.groupby('order_status').agg(count = ('order_status','count')).reset_index().sort_values(by='count', ascending=False)
print(status)
print(f"\n1. Most common = {status['order_status'].iloc[0]}")
print(f"\n1. Leas common = {status['order_status'].iloc[-1]}")
```

## Question 6

```python
merged_order_table = pd.merge(orders, order_items, on='order_id', how='left')
print(f"1. Total Rows after merge = {merged_order_table.shape[0]}")
print(f"2. Because in a left join, it not only retains all the rows from the previous df but also adds the relevant rows form the new df")
## Correction: One order could have multiple items, so the number of rows increased after the merge. One to Many join
```

## Question 7

```python
items_merge = pd.merge(order_items, products, on='product_id', how='left')
items_merge.groupby('product_category_name').agg(total_revenue=('price','sum')).reset_index().sort_values(by='total_revenue', ascending=False).iloc[:10]
```

## Question 8

```python
customer_orders = customers.merge(orders, on='customer_id', how='left').merge(order_items, on='order_id', how='left')
customer_orders.groupby('customer_unique_id').agg(total_spent = ('price','sum')).reset_index().sort_values(by='total_spent',ascending=False).iloc[:20]
```
